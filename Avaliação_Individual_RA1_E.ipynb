5. Juros compostos em investimentos:

Um investidor aplica R$ 5.000 em uma conta de poupança com uma taxa de juros anual de 6% composta mensalmente. Qual será o saldo da conta após 10 anos? Use a fórmula:

$$
A(t) = P \cdot (1 + \frac{r}{n})^{n \cdot t}
$$

onde $A(t)$ é o saldo após $t$ anos, $P$ é o principal investido, $r$ é a taxa de juros anual, $n$ é o número de vezes que os juros são compostos por ano e $t$ é o tempo em anos. Não se esqueça de traçar um gráfico mostrando os juros, ano a ano nos próximos 20 anos.


```python
#Nicolas Andreas Jackel - BCC - Turma U

import matplotlib.pyplot as plt
import math

#Variaveis
P = 5000  #Principal Investido
r = 0.06  #Taxa de Juros
t = 20  #Anos

#Calcula Saldo após 20 anos

Saldo = [P * math.pow(1 + r, i) for i in range(t + 1)] 
"""
onde i = n -> número de
vezes que os juros são
compostos por ano 
"""

print(f"A(t) = {P} * (1 + {r}) ^ {t}")
# Plot the results
plt.plot(range(t + 1), Saldo)
plt.xlabel('Anos')
plt.ylabel('Saldo')
plt.title('Juros')
plt.show()
```

Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $4^{x+2} - 2^{x+1} = 12$

2. $\log_{4}(x^2 - 4) - 2\log_{4}(x - 2) = 0$

```python
#Nicolas Andreas Jackel - BCC - Turma U

import numpy as np
import matplotlib.pyplot as plt
import math

#Definir a funcao
def f(x):
    return (4**x+2) - (2**x+1) - 12

valores_x = np.linspace(-2, 2, 1000)

valores_y = f(valores_x)

#Crescente ou Decrescente
if all(valores_y[i] < valores_y[i+1] for i in range(len(valores_y)-1)):
    trend = "Decrescente"
else:
    trend = "crescente"

print("1.")
# Plot the function
plt.plot(valores_x, valores_y)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f"Grafico da Funcao ({trend})")
plt.show()


print(trend)

```

# Questão 3
Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática, o passo-a-passo algébrico que leva a determinação dos indicados a seguir. Ao final, para cada uma das equações, você deverá traçar um gráfico da função. 

1. $\lim_{x \to 0} \frac{e^{2x} - 1}{\sin(x)}$

2. $\lim_{x \to 0} \frac{\tan(x) - x}{x^3}$

```python
#Nicolas Andreas Jackel - BCC - Turma U

import math
import matplotlib.pyplot as plt
import numpy as np

#Limite da funcao 1
def limite_de_1(x):
    if x == 0:
        return 2
    else:
        return (math.exp(2*x) - 1) / math.sin(x)

#Linite da funcao 2
def limite_de_2(x):
    if x == 0:
        return 1/3
    else:
        return (math.tan(x) - x) / (x**3)

valores_de_x = np.linspace(-1, 1, 1000)

valores_de_y1 = [limite_de_1(x) for x in valores_de_x]

valores_de_y2 = [limite_de_2(x) for x in valores_de_x]

#Primeio grafico
print("1.")
plt.plot(valores_de_x, valores_de_y1)
plt.title('Limite de 1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

#Segundo grafico
print("2.")
plt.plot(valores_de_x, valores_de_y2)
plt.title('Limite de 2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
